```python
import ccxt
import numpy as np
import pandas as pd
```

```python
exchange = ccxt.coinbasepro({'enableRateLimit': True})
candles = exchange.fetch_ohlcv('BTC/USD', '1d')
```

```python
df = pd.DataFrame(candles)
df.columns = ["Time", "Open", "Highest", "Lowest", "Closing", "Volume"]
```

```python
df.describe
```

```python
df['Time'] = pd.to_datetime(df['Time'], unit='ms')
```

```python
ax = df.plot(
    x='Time',
    y=['Open', 'Closing', 'Highest', 'Lowest'],
    title ="BTC/USD",
    figsize=(15, 5),
    legend=True,
    fontsize=12)
```
